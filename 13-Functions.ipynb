```python
def args_sum(*args):
    return sum(args)
```

```python
args_sum(10,20,30,60)
```

```python
def args_example(*args):
    print(args)
```

```python
args_example(123,21,13,41)
```

```python
def kwargs_example(**kwargs):
    print(kwargs)
```

```python
kwargs_example(apple = 100 , banana = 150 , melon = 200)
```
