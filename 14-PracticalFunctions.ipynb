```python
def divideNumber(number):
    return number / 2
```

```python
divideNumber(20)
```

```python
myList = [3,5,7,9,10,20,30]
```

```python
myResultList = []
for num in myList:
    myResultList.append(divideNumber(num))
print(myResultList)
```

# Map

```python
list(map(divideNumber,myList))
```

```python
def controlString(string):
    return "Halit" in string
```

```python
controlString("Halit İskitoğlu")
```

```python
myStringList = ["Halit","Halit Berk","Halit İskitoğlu","Berk"]
```

```python
list(map(controlString,myStringList))
```

# Filter

```python
list(filter(controlString,myStringList))
```

# LEGB  => L = Local , E => Enclosing , G => Global , B => Built-In

```python
#Global
myString = "Halit"
def myFunction():
    # Enclosing
    myString = "Halit 2"
    print(myString)
    def myFunction2():
        #Local
        myString = "Halit 3"
        print(myString)

    myFunction2()
```

```python
myString
```

```python
myFunction()
```

```python
myString
```

```python
y = 10
def newFunction(y):
    print(y)
    y = 5
    print(y)
    return y
```

```python
newFunction(10)
```

```python
y = 10
def changeY():
    global y
    y = 5
    print(y)
```
